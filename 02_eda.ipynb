```python
plt.figure(figsize=(12,10))
sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    annot_kws={"size":6}   # smaller text
)
plt.title("Correlation Heatmap")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
corr = df.corr(numeric_only=True)['ad_revenue_usd'] \
            .sort_values(ascending=False) \
            .round(2)

corr_df = corr.reset_index()
corr_df.columns = ['Feature', 'Correlation']

print(corr_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,4))

ax = sns.heatmap(
    corr[['ad_revenue_usd']].sort_values(by='ad_revenue_usd', ascending=False).T,
    annot=True,
    fmt=".2f",
    cmap='coolwarm'
)

# 🔁 Rotate annotation text (values)
for text in ax.texts:
    text.set_rotation(90)

plt.title("Correlation with Ad Revenue")
plt.tight_layout()
plt.show()
```
